```python
import pandas as pd

# Предположим, что у вас есть три DataFrame: df1, df2, df3

# Пример данных
data1 = {'ИНН': ['111', '222', '333'],
         'Дата': ['2024-04-16', '2024-04-16', '2024-04-16'],
         'Проверка1': ['Успех', 'Неудача', 'Успех']}
data2 = {'ИНН': ['111', '111', '111', '222', '333'],
         'Дата': ['2024-04-16', '2024-04-16', '2024-04-16', '2024-04-16', '2024-04-16'],
         'Контрагент': ['1', '2', '3', '1', '1'],
         'Проверка2': ['Триггер', 'Корректно', 'Триггер', 'Корректно', 'Корректно']}
data3 = {'ИНН': ['111', '222', '333'],
         'Дата': ['2024-04-16', '2024-04-16', '2024-04-16'],
         'Проверка3': ['Успех', 'Успех', 'Неудача']}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

# Объединение DataFrame по полям ИНН и ДАТА
merged_df = pd.merge(df1, df2, on=['ИНН', 'Дата'], how='outer')
merged_df = pd.merge(merged_df, df3, on=['ИНН', 'Дата'], how='outer')
# Создание нового поля 'ВЫВОД' с данными всех проверок
merged_df['ВЫВОД'] = merged_df[['Проверка1', 'Проверка2', 'Проверка3']].agg(', '.join, axis=1)
merged_df.drop(['Проверка1', 'Проверка2', 'Проверка3'], axis=1, inplace=True)

# Вывод нового DataFrame
print(merged_df)
```

```python
import pandas as pd

# Пример данных
data1 = {'ИНН': ['111', '222', '333'],
         'Дата': ['2024-04-16', '2024-04-16', '2024-04-16'],
         'Проверка1_1': ['Триггер', 'Корректно', 'Триггер'],
        'Проверка1_2': ['Триггер', 'Корректно', 'Триггер'],
        'Проверка1_3': ['Триггер', 'Корректно', 'Триггер']}
data2 = {'ИНН': ['111', '111', '111', '222', '333'],
         'Дата': ['2024-04-16', '2024-04-16', '2024-04-16', '2024-04-16', '2024-04-16'],
         'Контрагент': ['1', '2', '3', '1', '1'],
         'Проверка2': ['Триггер', 'Корректно', 'Триггер', 'Корректно', 'Корректно']}
data3 = {'ИНН': ['111', '222', '333'],
         'Дата': ['2024-04-16', '2024-04-16', '2024-04-16'],
         'Проверка3': ['Триггер', 'Триггер', 'Корректно']}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

# Объединение DataFrame по полям ИНН и ДАТА
merged_df = pd.merge(df1, df2, on=['ИНН', 'Дата'], how='outer')
print(merged_df)
```

```python
merged_df = pd.merge(merged_df, df3, on=['ИНН', 'Дата'], how='outer')

print(merged_df)
```

```python
# Создание нового поля 'ВЫВОД' с данными всех проверок
merged_df['ВЫВОД'] = merged_df[['Проверка1_1', 'Проверка1_2', 'Проверка1_3', 'Проверка2', 'Проверка3']].agg(', '.join, axis=1)

print(merged_df)
```

```python
merged_df.drop(['Проверка1_1', 'Проверка1_2', 'Проверка1_3', 'Проверка2', 'Проверка3'], axis=1, inplace=True)

print(merged_df)
```

```python
# Группировка по ИНН и ДАТА, объединение строк для каждой группы
merged_df = merged_df.groupby(['ИНН', 'Дата'], as_index=False).agg({'ВЫВОД': ', '.join})

```

```python
from tabulate import tabulate

# Преобразование DataFrame в формат таблицы
table = tabulate(merged_df, headers='keys', tablefmt='grid')

# Вывод таблицы
print(table)
```
