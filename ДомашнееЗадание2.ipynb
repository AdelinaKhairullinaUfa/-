```python
import pandas as pd
df = pd.read_csv('СМАД Тема 3 ДЗ 2.csv')
df
```

```python
import numpy as np
missing_values = df.isin(['?', np.nan]).sum()
print("Пропущенные значения:")
print(missing_values)
```

```python
# Замена пропущенных значений "?" на NaN
df.replace('?', np.nan, inplace=True)

# Удаление строк с пропущенными значениями
df.dropna(inplace=True)

# Преобразование категориальных переменных с помощью One-Hot Encoding
df = pd.get_dummies(df, columns=['income'])
df
```

```python
print(df.columns)
```

```python
education_grouped_income = df.groupby('education').agg({'income_<=50K': 'mean', 'income_>50K': 'mean'})
print("Средний доход для каждой группы образования:")
print(education_grouped_income)

# Изучим распределение целевой переменной "income"
income_distribution = df['income_<=50K'].value_counts(normalize=True)
print("\nРаспределение целевой переменной income:")
print(income_distribution)

# Рассчитаем статистические показатели для каждой колонки
column_statistics = df.describe()
print("\nСтатистика для каждой колонки:")
print(column_statistics)

# Найдем группу с максимальным и минимальным доходом
max_income_group = education_grouped_income['income_>50K'].idxmax()
min_income_group = education_grouped_income['income_>50K'].idxmin()

max_income_value = education_grouped_income.loc[max_income_group, 'income_>50K']
min_income_value = education_grouped_income.loc[min_income_group, 'income_>50K']

print("\nГруппа с максимальным доходом:", max_income_group, "(", max_income_value, ")")
print("Группа с минимальным доходом:", min_income_group, "(", min_income_value, ")")
df
```
